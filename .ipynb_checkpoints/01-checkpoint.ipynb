```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("pokemon.csv")
```

```python
df.head()
```

```python
df.info()
```

## To simplify the analysis and avoid the complexity of dealing with Pokemon having combined types, we will only consider Pokemon with a single type. In other words, we will exclude the Pokemon that have more than one type (dual types) from our analysis.

```python
df = df[df["type2"].isnull()].loc[: ,['sp_attack', 'sp_defense', 'attack', 'defense', 'speed', 'hp', 'type1']]
df.head()
```

```python
df = df.reset_index(drop=True)
df.head()
```

```python
X = df.drop("type1", axis=1)
y = df["type1"]
```

```python
from sklearn.preprocessing import normalize
X_norm = normalize(X)
```

## When we're working with data, sometimes we need to make the values easier to understand and compare. Normalization helps us do that by making the bigger values smaller and the smaller values bigger. It makes it easier for us to see what's going on and make sense of the data. However, we're utilizing normalization just for heatmap in order to see the correlation between the types not for LDA graph!

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis(n_components=3)
lda.fit(X_norm, y)
```

```python
lda.classes_
```

```python
lda.coef_
```

```python
coor_ = pd.DataFrame(lda.coef_, 
                         columns=df.columns[:-1], 
                         index=[lda.classes_])
```

```python
coor_
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(10,8))
sns.heatmap(data=coor_, annot=True, cmap="viridis", linewidths=0.5)
plt.suptitle('LDA Feature Coefficients')
```

## In this analysis, we're looking at how different classes of Pokemon are separated based on their attributes. We found that some classes, like rock, ghost, and fighting, have clear and distinct characteristics that set them apart from other classes. 
## On the other hand, ice and water classes have less distinctive attributes and are more difficult to distinguish from each other. 
## We can see this by looking at the coefficients that describe each class, and comparing the absolute total of coefficients for each class.

```python
lda.explained_variance_ratio_
```

```python
sum_of_rows = coor_.abs().sum(axis=1).sort_values()
```

```python
sorted_list = []
for row in sum_of_rows.index:
    for item in row:
        sorted_list.append(item)
        
print(sorted_list)
```

```python
sum_of_rows.values
```

```python
plt.figure(figsize=(12,5))
plt.bar(x=range(1,len(sum_of_rows)+1), height=sum_of_rows.values,
        tick_label=sorted_list)
plt.xticks(rotation=90)
plt.show()
```

## The analysis shows that there is a clear difference in separability between the rock and ghost categories compared to the water and ice categories, as evident from both the heatmap and bar plot. 
## However, it's important to note that the y-values in both plots are relative and lack an intuitive interpretation. While higher values imply better performance, the exact numbers themselves don't carry much meaning. 
## In order to better understand the performance differences and overall model performance, we need to apply Linear Discriminant Analysis (LDA).

```python
X = lda.fit_transform(X, y)
X
```

```python
len(X)
```

```python
labels = ["LDA" + str(x) for x in range(1, len(lda.explained_variance_ratio_)+1)]
labels
```

```python
X
```

```python
y
```

```python
lda_df = pd.DataFrame(X, columns=labels )
lda_df.head()
```

```python
lda_df["type"] = y.values
```

```python
lda_df.head()
```

```python
# For 2 LDA components:
plt.figure(figsize=(15,8))
sns.scatterplot(x="LDA1", y="LDA2", 
                data=lda_df, hue="type", 
                s=200, style="type")
plt.legend(bbox_to_anchor=(1.2, 1))
plt.show()
```

## The scatter plot we're looking at suggests that classifying Pokemon based solely on their stats is a non-linear problem, meaning it's difficult to separate them into distinct groups using just these features. This tells us that predicting a Pokemon's type based on stats alone is a challenging classification problem.

## Let's do for 3 LDA components:

```python
lda_df['type'].unique()
```

```python
from mpl_toolkits.mplot3d import Axes3D

colors = {'bug':'red', 'dark':'blue', 'dragon':'green', 'electric':'purple',
          'fairy':'orange', 'fighting':'black', 'fire':'gray', 'flying':'brown',
          'ghost':'pink', 'grass':'yellow', 'ground':'olive', 'ice':'cyan',
          'normal':'magenta', 'poison':'teal', 'psychic':'navy', 'rock':'khaki',
          'steel':'coral', 'water':'turquoise'}

c = lda_df['type'].apply(lambda x: colors[x])

fig = plt.figure(figsize=(15, 8))
ax = fig.add_subplot(111, projection="3d")

ax.scatter(xs=lda_df.LDA1, ys=lda_df.LDA2, zs=lda_df.LDA3, c=c, s=75)
ax.set_xlabel("LDA1")
ax.set_ylabel("LDA2")
ax.set_zlabel("LDA3")

import matplotlib.patches as mpatches

patches = [mpatches.Patch(color=color, label=label) for label, color in colors.items()]
plt.legend(handles=patches, bbox_to_anchor=(1.5, 0.8))
plt.show()
```

## For 3d graph, it is also hard to seperate them into distinct groups using just these features
